# Hierarchical Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("crime_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(df.iloc[:,1:])
```

```python
df_norm
```

```python
import scipy.cluster.hierarchy as sch #Dendogram
from sklearn.cluster import AgglomerativeClustering #clustering algorithm
```

```python
# create dendrogram

dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
df_norm['h_clusterid'] = Clusters
df_norm
```

```python
df_norm.sort_values("h_clusterid")
```

```python
df_norm.h_clusterid.value_counts()
```

# K-Means clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
df2 = pd.read_csv("crime_data.csv")
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(df.iloc[:,1:])
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
df2['Cluster']=x
```

```python
df2
```

```python
df2.iloc[:,1:5].groupby(df2.Cluster).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df3=pd.read_csv('crime_data.csv')
```

```python
df3.info()
```

```python
ps=df3.iloc[:,1:5]
```

```python
ps.values
```

```python
stscaler=StandardScaler().fit(ps.values)
x=stscaler.transform(ps.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
psl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
psl
```

```python
pd.concat([df3,psl],axis=1)
```
